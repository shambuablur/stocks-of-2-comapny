```
import yfinance as yf
RELIANCE.NS = yf.download('RELIANCE.NS',start='2015-07-05', end='2020-08-15')
RELIANCE.NS
```

```
import yfinance as yf
WIPRO.NS=yf.download('WIPRO.NS',start='2015-07-05', end='2020-08-15')
WIPRO.NS
```

```
import pandas as pd
data = [RELIANCE.NS['Close'],WIPRO.NS['Close']]
data = pd.DataFrame(data).transpose()
data = pd.DataFrame(data.values)
data = data.rename(columns={0:"RELIANCE",1 : "WIPRO" })
data
```

```
import matplotlib.pyplot as grp
data.plot(figsize=(10, 7))
grp.legend()
grp.title("Adjusted Close Price (15/08/2020) of RELNCE AND WIPRO companies ", fontsize=16)
grp.ylabel('Price', fontsize=14)
grp.xlabel('Year', fontsize=14)
grp.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
grp.show()
```

```
df = data 
grp.matshow(df.corr())
grp.xticks(range(len(df.columns)), df.columns)
grp.yticks(range(len(df.columns)), df.columns)
grp.colorbar()
grp.show()
```

```
RELIANCE.NS['Volume'].plot(figsize=(16,8),label = "RELIANCE",title='Volume')
WIPRO.NS['Volume'].plot(label = "WIPRO")
grp.legend();
```

```
RELIANCE.NS['Total Traded'] = RELIANCE.NS['Open'] * RELIANCE.NS['Volume']
WIPRO.NS['Total Traded'] = WIPRO.NS['Open'] * WIPRO.NS['Volume']
RELIANCE.NS['Total Traded'].plot(figsize=(16,8),label = "RELIANCE",title='Total Money Traded')
WIPRO.NS['Total Traded'].plot(label = "WIPRO")
grp.legend()
grp.ylabel('Total Money Traded')
```

```
RELIANCE.NS['returns'] = RELIANCE.NS['Close'].pct_change(1)
WIPRO.NS['returns'] = WIPRO.NS['Close'].pct_change(1)
```

```
fig = grp.figure(figsize=(10,5))
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax1.plot(RELIANCE.NS['returns'][-15:-5])
ax1.set_xlabel("Date")
ax1.set_ylabel("Percent")
ax1.set_title("RELNCE last month returns data")
grp.show()
```

```
fig = grp.figure(figsize=(10,5))
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax1.plot(WIPRO.NS['returns'][-15:-5])
ax1.set_xlabel("Date")
ax1.set_ylabel("Percent")
ax1.set_title("WIPRO last month returns data")
grp.show()
```

```
RELIANCE.NS['returns'].hist(bins=100)
```

```
WIPRO.NS['returns'].hist(bins=100)
```

```
RELIANCE.NS['returns'].hist(bins=50,label='RELNCE',figsize = (10,8), alpha= 0.5)
WIPRO.NS['returns'].hist(bins=50,label='WIPRO', alpha= 0.5)
grp.xlabel('Returns')
grp.legend();
```

```
RELIANCE.NS['daily_cumulative_return'] = ( 1 + RELIANCE.NS['returns'] ).cumprod()
WIPRO.NS['daily_cumulative_return'] = ( 1 + WIPRO.NS['returns'] ).cumprod()
RELIANCE.NS['daily_cumulative_return'].plot(label='RELIANCE',figsize = (16,8),title = 'Cumilative Return')
WIPRO.NS['daily_cumulative_return'].plot(label='WIPRO')
grp.legend();
```

```
"## RELIANCE"""

print ("RELIANCE.NS returns mean = ", RELIANCE.NS['returns'].mean(),"\n\n",
       "RELIANCE.NS returns mean * 250 = ",RELIANCE.NS['returns'].mean()*250,"\n\n",
       "RELIANCE.NS returns Standard Deviation = ",RELIANCE.NS['returns'].std(),"\n\n",
       "RELIANCE.NS returns Standard Deviation * 250 * 0.5 = ",RELIANCE.NS['returns'].std()*250**0.5)
```

```
"## RELIANCE"""

print ("WIPRO.NS returns mean = ", WIPRO.NS['returns'].mean(),"\n\n",
       "WIPRO.NS returns mean * 250 = ",WIPRO.NS['returns'].mean()*250,"\n\n",
       "WIPRO.NS returns Standard Deviation = ",WIPRO.NS['returns'].std(),"\n\n",
       "WIPRO.NS returns Standard Deviation * 250 * 0.5 = ",WIPRO.NS['returns'].std()*250**0.5)
```
